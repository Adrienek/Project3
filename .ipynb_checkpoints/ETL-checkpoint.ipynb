```python
import pandas as pd
import sqlite3
import requests
import json
import numpy as np
import pprint
```

```python
cities = {
    'atlanta':[33.749,-84.388],
    'philadelphia':[39.952583,-75.165222],
    'chicago':[41.881832,-87.623177],
    'houston':[29.7604,-95.3698],
    'seattle':[47.6062,-122.3321],
    'las_vegas':[36.1716,-115.1391]
}
```

```python
city_weather = {}
for city, coordinates in cities.items():
    lat = coordinates[0]
    lon = coordinates[1]
    url = (f"https://archive-api.open-meteo.com/v1/archive?latitude={lat}&longitude={lon}&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago")
    weather = requests.get(url).json()
    city_weather[city] = weather
    print(f"city: {city}, lat: {lat}, lon: {lon}")
    print("API URL:", url)
```

```python
atl_url = (f"https://archive-api.open-meteo.com/v1/archive?latitude=33.749&longitude=-84.388&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago")

atl_weather = requests.get(atl_url).json()

phil_url = (f"https://archive-api.open-meteo.com/v1/archive?latitude=39.952583&longitude=-75.165222&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago")

phil_weather = requests.get(phil_url).json()

chi_url = (f"https://archive-api.open-meteo.com/v1/archive?latitude=41.881832&longitude=-87.623177&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago")

chi_weather = requests.get(chi_url).json()
```

```python
atl_date = atl_weather['daily']['time']
atl_weather_code = atl_weather['daily']['weathercode']
atl_max_temp = atl_weather['daily']['temperature_2m_max']
atl_precip_sum = atl_weather['daily']['precipitation_sum']
atl_rain_sum = atl_weather['daily']['rain_sum']
atl_snow_sum = atl_weather['daily']['snowfall_sum']
atl_precip_hours = atl_weather['daily']['precipitation_hours']
atl_max_windspeed = atl_weather['daily']['windspeed_10m_max']

phil_date = phil_weather['daily']['time']
phil_weather_code = phil_weather['daily']['weathercode']
phil_max_temp = phil_weather['daily']['temperature_2m_max']
phil_precip_sum = phil_weather['daily']['precipitation_sum']
phil_rain_sum = phil_weather['daily']['rain_sum']
phil_snow_sum = phil_weather['daily']['snowfall_sum']
phil_precip_hours = phil_weather['daily']['precipitation_hours']
phil_max_windspeed = phil_weather['daily']['windspeed_10m_max']

chi_date = chi_weather['daily']['time']
chi_weather_code = chi_weather['daily']['weathercode']
chi_max_temp =chi_weather['daily']['temperature_2m_max']
chi_precip_sum = chi_weather['daily']['precipitation_sum']
chi_rain_sum = chi_weather['daily']['rain_sum']
chi_snow_sum = chi_weather['daily']['snowfall_sum']
chi_precip_hours = chi_weather['daily']['precipitation_hours']
chi_max_windspeed = chi_weather['daily']['windspeed_10m_max']
```

```python
atl_date
```

```python
columns = [
    "city_id","date","weather_code","precipitation_hours","precipitation_sum","rain_sum","snow_sum","max_temp","max_wind_speed"
]

atl_df = pd.DataFrame(columns=columns)
phil_df = pd.DataFrame(columns=columns)
chi_df = pd.DataFrame(columns=columns)
```

```python
atl_df
```

```python
atl_df['city_id']=[1]*365
atl_df['date']=atl_date
atl_df['weather_code']=atl_weather_code
atl_df['precipitation_hours']=atl_precip_hours
atl_df['precipitation_sum']=atl_precip_sum
atl_df['rain_sum']=atl_rain_sum
atl_df['snow_sum']=atl_snow_sum
atl_df['max_temp']=atl_max_temp
atl_df['max_wind_speed']=atl_max_windspeed

phil_df['city_id']=[2]*365
phil_df['date']=phil_date
phil_df['weather_code']=phil_weather_code
phil_df['precipitation_hours']=phil_precip_hours
phil_df['precipitation_sum']=phil_precip_sum
phil_df['rain_sum']=phil_rain_sum
phil_df['snow_sum']=phil_snow_sum
phil_df['max_temp']=phil_max_temp
phil_df['max_wind_speed']=phil_max_windspeed

chi_df['city_id']=[3]*365
chi_df['date']=chi_date
chi_df['weather_code']=chi_weather_code
chi_df['precipitation_hours']=chi_precip_hours
chi_df['precipitation_sum']=chi_precip_sum
chi_df['rain_sum']=chi_rain_sum
chi_df['snow_sum']=chi_snow_sum
chi_df['max_temp']=chi_max_temp
chi_df['max_wind_speed']=chi_max_windspeed
```

```python
atl_df['date']=pd.to_datetime(atl_df['date'])
phil_df['date']=pd.to_datetime(phil_df['date'])
chi_df['date']=pd.to_datetime(chi_df['date'])
```

```python
#export dataframe to sqlite file
filename = 'city_weather_info.db'
conn = sqlite3.connect(filename)
cur = conn.cursor()
```

```python
atl_df.to_sql('atl_weather',conn, if_exists='replace',index=False)
phil_df.to_sql('philly_weather',conn, if_exists='replace',index=False)
chi_df.to_sql('chi_weather',conn, if_exists='replace',index=False)
```

```python
cur.execute("CREATE TABLE if not exists combined_weather AS SELECT * FROM atl_weather UNION SELECT * FROM philly_weather UNION SELECT * FROM chi_weather ")
for row in cur:
    print(row)
```

```python
#creating fact table
cur.execute('''CREATE TABLE if not exists city_list (
           city_id INTEGER PRIMARY KEY,
            city_name TEXT,
            lat FLOAT,
            lon FLOAT)''')
conn.commit()
```

```python
new_rows = [(1, 'Atlanta', 33.7490, -84.3880),
            (2, 'Philadelphia', 39.9526, -75.1652),
           (3, 'Chicago', 41.881832,-87.623177)]
cur.executemany("INSERT INTO city_list (city_id, city_name, lat, lon) VALUES (?, ?, ?, ?)", new_rows)
conn.commit()
```
